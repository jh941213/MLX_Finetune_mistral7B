```python
#뉴스 데이터셋 

from datasets import load_dataset

dataset = load_dataset("daekeun-ml/naver-news-summarization-ko")
```

```python
dataset['train'][0]
```

```python
# 새로운 형식으로 데이터를 변환하는 코드
transformed_data = []
for item in dataset['train']:
    transformed_item = {"text": f"<s>[INST] Instruction : 당신은 뉴스기사를 요약하는 뉴스요약 봇입니다. 제목: {item['title']} 본문:{item['document']}[/INST] 요약:{item['summary']}</s>"}
    transformed_data.append(transformed_item)

# 결과 확인
for data in transformed_data[:1]:  # 예시 출력을 위해 첫 번째 항목만 출력
    print(data)
```

```python
from sklearn.model_selection import train_test_split
import json

# 데이터를 train, valid, test 세트로 8:1:1 비율로 나눕니다.
train_data, temp = train_test_split(transformed_data, test_size=0.2, random_state=42)
valid_data, test_data = train_test_split(temp, test_size=0.5, random_state=42)

# 각 데이터 세트를 jsonl 파일로 저장합니다.
with open('train.jsonl', 'w', encoding='utf-8') as f:
    for entry in train_data:
        json.dump(entry, f, ensure_ascii=False)
        f.write('\n')

with open('valid.jsonl', 'w', encoding='utf-8') as f:
    for entry in valid_data:
        json.dump(entry, f, ensure_ascii=False)
        f.write('\n')

with open('test.jsonl', 'w', encoding='utf-8') as f:
    for entry in test_data:
        json.dump(entry, f, ensure_ascii=False)
        f.write('\n')
```
